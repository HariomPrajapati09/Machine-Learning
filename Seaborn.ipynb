***Seaborn***

*   Data Visualization Library



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Note: Seaborn has some built-in datasets

```python
# Total bill vs tip datasets
tips=sns.load_dataset('tips')
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# Load the iris dataset
isis=sns.load_dataset('iris')
isis.head()
```

***Scatter plot***

```python
sns.scatterplot(data=isis,x='sepal_length',y='petal_length',hue='species')
```

```python
sns.scatterplot(data=isis,x='sepal_length',y='sepal_width',hue='species')
```

```python

# Loading the titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(data=titanic,x='class')
```

```python
sns.countplot(data=titanic,x='alive')
```

```python
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```
