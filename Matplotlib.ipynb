***Matplotlib:-***
*   **Useful for making Plots**



```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

***Plotting the data***

```python
#Sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#Adding title on x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

***Parabola***

```python
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
plt.show()
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'r*')
plt.plot(x,np.cos(x),'bx')
plt.show()
```

***Bar Plot***

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['Java','Python','PHP','JavaScript','C#','C++']
students=[23,17,35,29,12,20]
ax.bar(languages,students)

plt.show()
```

***Pie Chart***

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['Java','Python','PHP','JavaScript','C#','C++']
students=[23,17,35,29,12,20]
ax.pie(students,labels=languages,autopct='%1.1f%%')
plt.show()
```

***Scatter plot***

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```

***3D Scater Plot***

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
