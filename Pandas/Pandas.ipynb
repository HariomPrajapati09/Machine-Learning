***Manipulatinf dataframes***

**1. Adding a column to a dataframe:-**

**dataframe['columnName']=dataset.target**

**2. Removing a particular row:-**

**dataframe.drop(index=indexvalue,axis=0)**

**3. Removing a particular column:-**

**dataframe.drop(columns='columnName',axis=1)**

**4. Locationg a row using the index valur:-**

**dataframe.iloc[index]**

**5. Locationg a particular column using the index valur:-**

**dataframe.iloc[ : ,index]**
