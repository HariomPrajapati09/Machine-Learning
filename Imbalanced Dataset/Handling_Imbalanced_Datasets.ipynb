**Imbalanced Datasets**

A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
#Loading dataset
credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

**0 --> Legit Transactions**

**1 --> Fraudulent Transactions**

```python
# seperating legit and fraudulent transactions
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Under-sampling**

Building a sample dataset containing similar distribution of Legit and Fraudulent Transactions

```python
legit_sample=legit.sample(n=371)
```

```python
print(legit_sample.shape)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```
